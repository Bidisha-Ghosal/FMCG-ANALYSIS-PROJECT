```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ── Load from your Excel file ──────────────────────────────────────────
file = "FMCG_Companies_Combuned_Data.xlsx"

peer = pd.read_excel(file, sheet_name="Peer_Summary", skiprows=1)
peer.columns = ["Metric", "ITC", "Nestle", "HUL", "Dabur", "Notes"]
peer = peer.dropna(subset=["ITC"]).reset_index(drop=True)

ratios = pd.read_excel(file, sheet_name="Key_Ratios", skiprows=1)
ratios.columns = ["Ratio", "ITC", "Nestle", "HUL", "Dabur", "Formula"]
ratios = ratios.dropna(subset=["ITC"]).reset_index(drop=True)

# ── Company colors (consistent across all charts) ──────────────────────
COLORS = {
    "ITC":    "#534AB7",   # purple
    "Nestle": "#1D9E75",   # teal
    "HUL":    "#D85A30",   # coral
    "Dabur":  "#BA7517",   # amber
}
COMPANIES = ["ITC", "Nestle", "HUL", "Dabur"]

print("Peer Summary shape:", peer.shape)
print("Key Ratios shape:  ", ratios.shape)
print("\nAvailable metrics:\n", peer["Metric"].tolist())
```

```python
# REVENUE EBITDA BARCHART
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Revenue & EBITDA Comparison — FY 2021-22 (₹ Crores)",
             fontsize=14, fontweight="bold", y=1.02)

# ── Revenue ────────────────────────────────────────────────────────────
revenue_row = peer[peer["Metric"].str.contains("Revenue from Operations", na=False)]
revenue = [float(revenue_row[c].values[0]) for c in COMPANIES]

bars = axes[0].bar(COMPANIES, revenue,
                   color=[COLORS[c] for c in COMPANIES],
                   width=0.5, edgecolor="white", linewidth=0.8)
axes[0].set_title("Revenue from Operations", fontsize=12, pad=10)
axes[0].set_ylabel("₹ Crores")
axes[0].yaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
for bar, val in zip(bars, revenue):
    axes[0].text(bar.get_x() + bar.get_width()/2,
                 bar.get_height() + 300,
                 f"₹{val:,.0f}", ha="center", va="bottom",
                 fontsize=9, fontweight="bold")
axes[0].spines[["top","right"]].set_visible(False)

# ── EBITDA ─────────────────────────────────────────────────────────────
ebitda_row = peer[peer["Metric"].str.contains("EBITDA", na=False) &
                  ~peer["Metric"].str.contains("Margin", na=False)]
ebitda = [float(ebitda_row[c].values[0]) for c in COMPANIES]

bars2 = axes[1].bar(COMPANIES, ebitda,
                    color=[COLORS[c] for c in COMPANIES],
                    width=0.5, edgecolor="white", linewidth=0.8)
axes[1].set_title("EBITDA", fontsize=12, pad=10)
axes[1].set_ylabel("₹ Crores")
axes[1].yaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
for bar, val in zip(bars2, ebitda):
    axes[1].text(bar.get_x() + bar.get_width()/2,
                 bar.get_height() + 100,
                 f"₹{val:,.0f}", ha="center", va="bottom",
                 fontsize=9, fontweight="bold")
axes[1].spines[["top","right"]].set_visible(False)

plt.tight_layout()
plt.savefig("output/chart_revenue_ebitda.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# MARGINS COMPARISON
fig, ax = plt.subplots(figsize=(12, 5))

# ── Pull margin rows ───────────────────────────────────────────────────
ebitda_margin_row = peer[peer["Metric"].str.contains("EBITDA Margin", na=False)]
pat_margin_row    = peer[peer["Metric"].str.contains("PAT Margin",    na=False)]
yoy_row           = peer[peer["Metric"].str.contains("YoY Revenue",   na=False)]

ebitda_margins = [float(ebitda_margin_row[c].values[0]) * 100 for c in COMPANIES]
pat_margins    = [float(pat_margin_row[c].values[0])    * 100 for c in COMPANIES]
yoy_growth     = [float(yoy_row[c].values[0])           * 100 for c in COMPANIES]

x = np.arange(len(COMPANIES))
width = 0.25

b1 = ax.bar(x - width, ebitda_margins, width, label="EBITDA Margin %",
            color=[COLORS[c] for c in COMPANIES], alpha=0.9, edgecolor="white")
b2 = ax.bar(x,         pat_margins,    width, label="PAT Margin %",
            color=[COLORS[c] for c in COMPANIES], alpha=0.55, edgecolor="white")
b3 = ax.bar(x + width, yoy_growth,     width, label="YoY Revenue Growth %",
            color=[COLORS[c] for c in COMPANIES], alpha=0.3,
            edgecolor=[COLORS[c] for c in COMPANIES], linewidth=1.2)

# Value labels
for bars in [b1, b2, b3]:
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.3,
                f"{bar.get_height():.1f}%",
                ha="center", va="bottom", fontsize=8)

ax.set_title("Margin & Growth Comparison — FY 2021-22",
             fontsize=13, fontweight="bold", pad=12)
ax.set_ylabel("Percentage (%)")
ax.set_xticks(x)
ax.set_xticklabels(COMPANIES, fontsize=11)
ax.spines[["top","right"]].set_visible(False)

# Legend using company colors + shade for metric type
legend_handles = [
    mpatches.Patch(color="#555", alpha=0.9, label="EBITDA Margin %"),
    mpatches.Patch(color="#555", alpha=0.55, label="PAT Margin %"),
    mpatches.Patch(facecolor="white", edgecolor="#555", linewidth=1.2,
                   label="YoY Revenue Growth %"),
]
ax.legend(handles=legend_handles, loc="upper right", fontsize=9)

plt.tight_layout()
plt.savefig("output/chart_margins.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# KEY RATIOS HEATMAP

fig, ax = plt.subplots(figsize=(10, 6))

# ── Select ratio rows of interest ─────────────────────────────────────
ratio_labels = [
    "Revenue Growth YoY (%)",
    "EBITDA Margin (%)",
    "PAT Margin (%)",
    "Return on Equity (%)",
    "Return on Assets (%)",
]
# filter by partial match
filtered = ratios[ratios["Ratio"].str.contains(
    "Revenue Growth|EBITDA Margin|PAT Margin|Return on Equity|Return on Assets",
    na=False, case=False
)]

matrix = filtered[COMPANIES].astype(float).values * 100  # convert decimals to %

im = ax.imshow(matrix, cmap="YlGn", aspect="auto")

# Labels
ax.set_xticks(range(len(COMPANIES)))
ax.set_xticklabels(COMPANIES, fontsize=12, fontweight="bold")
ax.set_yticks(range(len(ratio_labels)))
ax.set_yticklabels(ratio_labels, fontsize=10)

# Cell values
for i in range(len(ratio_labels)):
    for j in range(len(COMPANIES)):
        ax.text(j, i, f"{matrix[i, j]:.1f}%",
                ha="center", va="center",
                fontsize=11, fontweight="bold",
                color="white" if matrix[i, j] > matrix[i].mean() else "#333")

ax.set_title("Key Ratios Heatmap — FY 2021-22  (darker = stronger)",
             fontsize=13, fontweight="bold", pad=12)
plt.colorbar(im, ax=ax, shrink=0.8, label="Value (%)")
plt.tight_layout()
plt.savefig("output/chart_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# RADAR/ SPIDER CHART

from matplotlib.patches import FancyArrowPatch

categories = ["EBITDA\nMargin", "PAT\nMargin", "ROE", "ROA", "Revenue\nGrowth"]
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]   # close the polygon

# ── Values (normalise 0–1 within each metric for fair comparison) ──────
raw = {
    "ITC":    [35.96, 25.21, 24.53, 20.05, 23.12],
    "Nestle": [24.19, 14.59, 104.5, 26.13, 10.12],   # ROE >100 = asset-light model
    "HUL":    [24.99, 17.23, 18.08, 12.64, 11.29],
    "Dabur":  [25.37, 17.52, 24.43, 16.68, 13.88],
}

# Normalize each metric to 0–1
raw_arr = np.array(list(raw.values()))
col_max = raw_arr.max(axis=0)
col_max[col_max == 0] = 1
norm = {c: (np.array(v) / col_max).tolist() for c, v in raw.items()}

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for company in COMPANIES:
    vals = norm[company] + norm[company][:1]
    ax.plot(angles, vals, "o-", linewidth=2,
            color=COLORS[company], label=company)
    ax.fill(angles, vals, alpha=0.08, color=COLORS[company])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=11)
ax.set_yticklabels([])
ax.set_title("Relative Performance Radar\n(normalised — larger = better)",
             fontsize=13, fontweight="bold", pad=20)
ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1), fontsize=10)

plt.tight_layout()
plt.savefig("output/chart_radar.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
from scipy.stats import pearsonr
import os

# ── Inflation data ─────────────────────────────────────
inf_path = r"E:\FMCG_DATASET\CPI_DATA.csv"

df_inf = pd.read_csv(inf_path)

# Remove unwanted index column if present
df_inf = df_inf.loc[:, ~df_inf.columns.str.contains('^Unnamed')]

# Rename ALL 4 columns properly
df_inf.columns = ["Month", "Rural_CPI", "Urban_CPI", "Inflation_Gap"]

# ── Convert to numeric ────────────────────────────────
df_inf["Rural_CPI"] = pd.to_numeric(df_inf["Rural_CPI"], errors="coerce")
df_inf["Urban_CPI"] = pd.to_numeric(df_inf["Urban_CPI"], errors="coerce")
df_inf["Inflation_Gap"] = pd.to_numeric(df_inf["Inflation_Gap"], errors="coerce")

# ── Convert Month to datetime ─────────────────────────
df_inf["Date"] = pd.to_datetime(df_inf["Month"], format="%b-%y", errors="coerce")

# Drop invalid rows (if any)
df_inf = df_inf.dropna(subset=["Date"])

# Sort properly
df_inf = df_inf.sort_values("Date").reset_index(drop=True)

# ── Fiscal Year Function ──────────────────────────────
def to_fy(dt):
    return f"FY{dt.year}" if dt.month >= 4 else f"FY{dt.year - 1}"

df_inf["FY"] = df_inf["Date"].apply(to_fy)

# ── Output check ──────────────────────────────────────
print(df_inf.head(10))
print("\nFiscal years covered:", df_inf["FY"].unique())
```

```python
# ANNUAL FMCG PERFORMANCE TABLE

# ── Annual FMCG financials (from your cleaned data / notebook cells) ───
fmcg = pd.DataFrame({
    "Company":      ["ITC",      "ITC",      "HUL",      "HUL",
                     "Nestle",   "Nestle",   "Dabur",    "Dabur"],
    "FY":           ["FY2021",   "FY2022",   "FY2021",   "FY2022",
                     "FY2020",   "FY2021",   "FY2021",   "FY2022"],
    "Revenue":      [48524.56,   59745.56,   45996.00,   51193.00,
                     13390.03,   14709.41,    7184.73,    8179.50],
    "EBITDA_Margin":[0.3290,     0.3596,     0.2410,     0.2499,
                     0.2350,     0.2419,     0.2420,     0.2537],
    "PAT_Margin":   [0.2101,     0.2521,     0.1640,     0.1723,
                     0.1390,     0.1459,     0.1640,     0.1752],
    "Revenue_Growth":[np.nan,    0.2312,     np.nan,     0.1129,
                      np.nan,    0.1012,     np.nan,     0.1388],
})

print(fmcg)
```

```python
# ANNUAL AVERAGE CPI

# ── Annual average CPI per fiscal year ────────────────────────────────
annual_cpi = (df_inf.groupby("FY")
                    .agg(Avg_Rural_CPI  = ("Rural_CPI",  "mean"),
                         Avg_Urban_CPI  = ("Urban_CPI",  "mean"),
                         Avg_Gap        = ("Inflation_Gap","mean"))
                    .round(2)
                    .reset_index())

print("Annual CPI averages:\n", annual_cpi)

# ── Merge on FY ────────────────────────────────────────────────────────
merged = fmcg.merge(annual_cpi, on="FY", how="left")
merged["EBITDA_Pct"] = (merged["EBITDA_Margin"] * 100).round(2)
merged["PAT_Pct"]    = (merged["PAT_Margin"]    * 100).round(2)

print("\nMerged dataset:\n",
      merged[["Company","FY","Revenue","Avg_Urban_CPI",
              "EBITDA_Pct","PAT_Pct"]].to_string(index=False))
```

```python
# MONTHLY CPI TREND LINE CHART

COLORS = {
    "ITC":    "#534AB7",
    "Nestle": "#1D9E75",
    "HUL":    "#D85A30",
    "Dabur":  "#BA7517",
}

fig, ax = plt.subplots(figsize=(13, 5))

ax.plot(df_inf["Date"], df_inf["Rural_CPI"],
        color="#D85A30", linewidth=2, marker="o", markersize=4,
        label="Rural CPI (%)")
ax.plot(df_inf["Date"], df_inf["Urban_CPI"],
        color="#534AB7", linewidth=2, marker="s", markersize=4,
        label="Urban CPI (%)")
ax.fill_between(df_inf["Date"],
                df_inf["Rural_CPI"], df_inf["Urban_CPI"],
                alpha=0.12, color="#BA7517",
                label="Rural–Urban gap")

# Annotate peak
peak_idx = df_inf["Rural_CPI"].idxmax()
ax.annotate(f'Peak: {df_inf.loc[peak_idx,"Rural_CPI"]:.1f}%',
            xy=(df_inf.loc[peak_idx,"Date"],
                df_inf.loc[peak_idx,"Rural_CPI"]),
            xytext=(20, 10), textcoords="offset points",
            fontsize=9, arrowprops=dict(arrowstyle="->", color="#999"))

ax.set_title("Monthly CPI Trend — Rural vs Urban",
             fontsize=13, fontweight="bold", pad=10)
ax.set_ylabel("Inflation Rate (%)")
ax.set_xlabel("")
ax.legend(fontsize=10)
ax.spines[["top","right"]].set_visible(False)
ax.xaxis.set_major_formatter(
    plt.matplotlib.dates.DateFormatter("%b-%y"))
plt.xticks(rotation=30, ha="right")

plt.tight_layout()
plt.savefig("output/chart_cpi_trend.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# SCATTER: CPI vs EBITDA margin per company

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Inflation vs Profitability — FMCG Companies",
             fontsize=13, fontweight="bold")

companies = ["ITC", "Nestle", "HUL", "Dabur"]

for ax, (y_col, y_label) in zip(
        axes,
        [("EBITDA_Pct", "EBITDA Margin (%)"),
         ("PAT_Pct",    "PAT Margin (%)")]):

    for company in companies:
        sub = merged[(merged["Company"] == company) &
                     merged["Avg_Urban_CPI"].notna() &
                     merged[y_col].notna()]
        if len(sub) < 2:
            continue

        ax.scatter(sub["Avg_Urban_CPI"], sub[y_col],
                   color=COLORS[company], s=120, zorder=5,
                   label=company, edgecolors="white", linewidth=0.8)

        # Trend line
        z = np.polyfit(sub["Avg_Urban_CPI"], sub[y_col], 1)
        p = np.poly1d(z)
        x_line = np.linspace(sub["Avg_Urban_CPI"].min(),
                              sub["Avg_Urban_CPI"].max(), 50)
        ax.plot(x_line, p(x_line),
                color=COLORS[company], linewidth=1.2,
                linestyle="--", alpha=0.6)

        # Label points
        for _, row in sub.iterrows():
            ax.annotate(row["FY"],
                        (row["Avg_Urban_CPI"], row[y_col]),
                        textcoords="offset points",
                        xytext=(5, 5), fontsize=8,
                        color=COLORS[company])

    ax.set_xlabel("Avg Urban CPI (%)", fontsize=11)
    ax.set_ylabel(y_label, fontsize=11)
    ax.set_title(y_label, fontsize=11, pad=8)
    ax.spines[["top","right"]].set_visible(False)
    ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig("output/chart_cpi_vs_margins.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# CORRELATION HEATMAP: CPI VS OTHER MATRICS

fig, ax = plt.subplots(figsize=(9, 5))

metrics     = ["Revenue", "EBITDA_Pct", "PAT_Pct"]
cpi_vars    = ["Avg_Rural_CPI", "Avg_Urban_CPI", "Avg_Gap"]
metric_labels = ["Revenue (₹ Cr)", "EBITDA Margin %", "PAT Margin %"]
cpi_labels    = ["Rural CPI", "Urban CPI", "Rural–Urban Gap"]

corr_matrix = np.zeros((len(cpi_vars), len(metrics)))

for i, cpi in enumerate(cpi_vars):
    for j, metric in enumerate(metrics):
        sub = merged[[cpi, metric]].dropna()
        if len(sub) >= 2:
            r, _ = pearsonr(sub[cpi], sub[metric])
            corr_matrix[i, j] = round(r, 2)

im = ax.imshow(corr_matrix, cmap="RdYlGn",
               vmin=-1, vmax=1, aspect="auto")

ax.set_xticks(range(len(metrics)))
ax.set_xticklabels(metric_labels, fontsize=11)
ax.set_yticks(range(len(cpi_vars)))
ax.set_yticklabels(cpi_labels, fontsize=11)

for i in range(len(cpi_vars)):
    for j in range(len(metrics)):
        val = corr_matrix[i, j]
        ax.text(j, i, f"r = {val:.2f}",
                ha="center", va="center",
                fontsize=12, fontweight="bold",
                color="white" if abs(val) > 0.4 else "#333")

plt.colorbar(im, ax=ax, shrink=0.8, label="Pearson r")
ax.set_title("Correlation: CPI Inflation vs FMCG Financial Metrics",
             fontsize=13, fontweight="bold", pad=12)

plt.tight_layout()
plt.savefig("output/chart_corr_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.linear_model    import LinearRegression, Ridge
from sklearn.ensemble        import RandomForestRegressor, GradientBoostingRegressor
from sklearn.cluster         import KMeans
from sklearn.preprocessing   import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics         import (mean_absolute_error,
                                     mean_squared_error, r2_score)
from sklearn.decomposition   import PCA
import warnings
warnings.filterwarnings("ignore")

COLORS = {
    "ITC":    "#534AB7",
    "Nestle": "#1D9E75",
    "HUL":    "#D85A30",
    "Dabur":  "#BA7517",
}

# ── Master dataset: all companies, both years, all features ───────────
master = pd.DataFrame([
    # company, fy, revenue, ebitda_margin, pat_margin, roe, roa,
    # yoy_growth, cost_materials, rural_cpi, urban_cpi,
    # cereal_export_growth, sugar_export_growth
    ["ITC",    "FY21", 48524.56, 0.329, 0.210, 0.198, 0.175,  np.nan, 13605, 6.2,  5.8, 51.4, 15.3],
    ["ITC",    "FY22", 59745.56, 0.360, 0.252, 0.245, 0.201,  0.231,  16064, 7.1,  6.5, 92.9, 61.7],
    ["HUL",    "FY21", 45996.00, 0.241, 0.164, 0.162, 0.118,  np.nan, 14951, 6.2,  5.8, 51.4, 15.3],
    ["HUL",    "FY22", 51193.00, 0.250, 0.172, 0.181, 0.126,  0.113,  15869, 7.1,  6.5, 92.9, 61.7],
    ["Nestle", "FY21", 14709.41, 0.242, 0.146, 1.045, 0.261,  0.101,   5212, 6.2,  5.8, 51.4, 15.3],
    ["Nestle", "FY22", 14829.52, 0.242, 0.146, 1.045, 0.261,  0.082,   6105, 7.1,  6.5, 92.9, 61.7],
    ["Dabur",  "FY21",  7184.73, 0.242, 0.164, 0.231, 0.155,  np.nan,  2756, 6.2,  5.8, 51.4, 15.3],
    ["Dabur",  "FY22",  8179.50, 0.254, 0.175, 0.244, 0.167,  0.139,   3228, 7.1,  6.5, 92.9, 61.7],
],
columns=["Company","FY","Revenue","EBITDA_Margin","PAT_Margin",
         "ROE","ROA","YoY_Growth","Cost_Materials",
         "Rural_CPI","Urban_CPI",
         "Cereal_Export_Growth","Sugar_Export_Growth"])

# Encode company as number
le = LabelEncoder()
master["Company_Code"] = le.fit_transform(master["Company"])

# Derived features
master["Material_Intensity"] = (
    master["Cost_Materials"] / master["Revenue"] * 100).round(3)
master["CPI_Gap"]      = master["Rural_CPI"] - master["Urban_CPI"]
master["EBITDA_Pct"]   = (master["EBITDA_Margin"] * 100).round(2)
master["PAT_Pct"]      = (master["PAT_Margin"]    * 100).round(2)

print(master[["Company","FY","Revenue","EBITDA_Pct",
              "Material_Intensity","Urban_CPI"]].to_string(index=False))
```

```python
# LINEAR REGRESSION: PREDICTING REVENUE

# ── Features → Revenue ────────────────────────────────────────────────
features = ["Urban_CPI","Rural_CPI","Cost_Materials",
            "EBITDA_Margin","Cereal_Export_Growth",
            "Sugar_Export_Growth","Company_Code"]

df_lr = master[features + ["Revenue"]].dropna()

X = df_lr[features].values
y = df_lr["Revenue"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_s, y_train)
y_pred = lr.predict(X_test_s)

mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

print("── Model 1: Linear Regression ──────────────────────────")
print(f"  MAE  : ₹{mae:,.0f} Crores")
print(f"  RMSE : ₹{rmse:,.0f} Crores")
print(f"  R²   : {r2:.3f}")

# Feature importance via coefficients
coef_df = pd.DataFrame({
    "Feature":     features,
    "Coefficient": lr.coef_
}).sort_values("Coefficient", key=abs, ascending=False)
print("\n  Feature coefficients (standardised):")
print(coef_df.to_string(index=False))

# ── Plot: actual vs predicted ─────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(y_test, y_pred, color="#534AB7", s=120,
           edgecolors="white", linewidth=0.8, zorder=5)
lims = [min(y_test.min(), y_pred.min()) - 1000,
        max(y_test.max(), y_pred.max()) + 1000]
ax.plot(lims, lims, "k--", linewidth=1, alpha=0.5, label="Perfect fit")
for actual, predicted, idx in zip(
        y_test, y_pred,
        df_lr.index[-len(y_test):]):
    company = master.loc[idx, "Company"] if idx in master.index else ""
    ax.annotate(company, (actual, predicted),
                textcoords="offset points",
                xytext=(6, 4), fontsize=8, color="#534AB7")
ax.set_xlabel("Actual Revenue (₹ Cr)", fontsize=11)
ax.set_ylabel("Predicted Revenue (₹ Cr)", fontsize=11)
ax.set_title(f"Linear Regression — Actual vs Predicted Revenue\nR² = {r2:.3f}",
             fontsize=12, fontweight="bold", pad=10)
ax.legend(fontsize=10)
ax.spines[["top","right"]].set_visible(False)
plt.tight_layout()
plt.savefig("output/ml_lr_actual_vs_pred.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# RIDGE REGRESSION : PREDICTING EBITDA MARGIN

# ── Features → EBITDA Margin ──────────────────────────────────────────
features2 = ["Urban_CPI","Material_Intensity","Cereal_Export_Growth",
             "Sugar_Export_Growth","CPI_Gap","Company_Code"]

df_r = master[features2 + ["EBITDA_Pct"]].dropna()

X2 = df_r[features2].values
y2 = df_r["EBITDA_Pct"].values

scaler2 = StandardScaler()
X2_s    = scaler2.fit_transform(X2)

# Ridge with cross-validation
ridge = Ridge(alpha=1.0)
cv_scores = cross_val_score(ridge, X2_s, y2,
                             cv=3, scoring="r2")

ridge.fit(X2_s, y2)
y2_pred = ridge.predict(X2_s)

r2_ridge = r2_score(y2, y2_pred)
mae_r    = mean_absolute_error(y2, y2_pred)

print("── Model 2: Ridge Regression ───────────────────────────")
print(f"  CV R² scores : {cv_scores.round(3)}")
print(f"  Mean CV R²   : {cv_scores.mean():.3f}")
print(f"  Train R²     : {r2_ridge:.3f}")
print(f"  MAE          : {mae_r:.2f}%")

# ── Plot: feature coefficients ────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 4))
coef_r = pd.DataFrame({
    "Feature":     features2,
    "Coefficient": ridge.coef_
}).sort_values("Coefficient")

colors_bar = ["#D85A30" if c < 0 else "#1D9E75"
              for c in coef_r["Coefficient"]]
bars = ax.barh(coef_r["Feature"], coef_r["Coefficient"],
               color=colors_bar, alpha=0.85,
               edgecolor="white", height=0.5)
ax.axvline(0, color="#999", linewidth=0.8, linestyle="--")
for bar, val in zip(bars, coef_r["Coefficient"]):
    ax.text(val + (0.05 if val >= 0 else -0.05),
            bar.get_y() + bar.get_height()/2,
            f"{val:+.3f}", va="center", fontsize=9)
ax.set_title("Ridge Regression — Feature Impact on EBITDA Margin\n"
             "(green = increases margin, red = decreases)",
             fontsize=12, fontweight="bold", pad=10)
ax.set_xlabel("Standardised Coefficient", fontsize=11)
ax.spines[["top","right"]].set_visible(False)
plt.tight_layout()
plt.savefig("output/ml_ridge_coef.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# K-MEANS CLUSTERING: SEGMENTING COMPANIES BY PROFILE

# ── Cluster companies by financial profile ────────────────────────────
cluster_features = ["EBITDA_Pct","PAT_Pct","ROE","ROA",
                    "Material_Intensity","YoY_Growth"]

df_cl = master[["Company","FY"] + cluster_features].dropna()

scaler3 = StandardScaler()
X3 = scaler3.fit_transform(df_cl[cluster_features])

# ── Elbow method ──────────────────────────────────────────────────────
inertias = []
k_range  = range(2, 6)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X3)
    inertias.append(km.inertia_)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: elbow
ax = axes[0]
ax.plot(list(k_range), inertias, "o-",
        color="#534AB7", linewidth=2.2, markersize=8)
ax.set_title("Elbow Method — Optimal K", fontsize=11, pad=8)
ax.set_xlabel("Number of clusters (K)")
ax.set_ylabel("Inertia")
ax.spines[["top","right"]].set_visible(False)

# ── Fit K=2 (two clear profiles: high-margin vs mass-market) ─────────
km_final = KMeans(n_clusters=2, random_state=42, n_init=10)
df_cl["Cluster"] = km_final.fit_predict(X3)

cluster_labels = {0: "High-margin", 1: "Volume-driven"}
df_cl["Profile"] = df_cl["Cluster"].map(cluster_labels)

# Right: scatter — EBITDA% vs Material Intensity, colored by cluster
ax2 = axes[1]
cluster_colors = ["#534AB7", "#D85A30"]

for c_id, profile in cluster_labels.items():
    sub = df_cl[df_cl["Cluster"] == c_id]
    ax2.scatter(sub["Material_Intensity"], sub["EBITDA_Pct"],
                color=cluster_colors[c_id], s=140, zorder=5,
                label=profile,
                edgecolors="white", linewidth=0.8)
    for _, row in sub.iterrows():
        ax2.annotate(f'{row["Company"]} {row["FY"]}',
                     (row["Material_Intensity"], row["EBITDA_Pct"]),
                     textcoords="offset points",
                     xytext=(6, 4), fontsize=8,
                     color=cluster_colors[c_id])

ax2.set_xlabel("Material Intensity (Cost / Revenue %)", fontsize=11)
ax2.set_ylabel("EBITDA Margin (%)", fontsize=11)
ax2.set_title("K-Means Clusters — Company Financial Profiles",
              fontsize=11, pad=8)
ax2.legend(fontsize=10)
ax2.spines[["top","right"]].set_visible(False)

plt.suptitle("K-Means Clustering — FMCG Company Profiles",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("output/ml_kmeans_clusters.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nCluster assignments:")
print(df_cl[["Company","FY","Profile","EBITDA_Pct",
             "Material_Intensity"]].to_string(index=False))
```

```python
# RANDOM FOREST : PREDICTING REVENUE GROWTH

# ── Features → YoY Revenue Growth ─────────────────────────────────────
features4 = ["Urban_CPI","Rural_CPI","Material_Intensity",
             "EBITDA_Margin","Cereal_Export_Growth",
             "Sugar_Export_Growth","Company_Code"]

df_rf = master[features4 + ["YoY_Growth"]].dropna()

X4 = df_rf[features4].values
y4 = df_rf["YoY_Growth"].values

# Note: small dataset — use all for training, report CV score
scaler4 = StandardScaler()
X4_s    = scaler4.fit_transform(X4)

rf = RandomForestRegressor(n_estimators=100,
                            max_depth=3,
                            random_state=42)
cv_rf = cross_val_score(rf, X4_s, y4, cv=3, scoring="r2")
rf.fit(X4_s, y4)

print("── Model 4: Random Forest ──────────────────────────────")
print(f"  CV R² scores : {cv_rf.round(3)}")
print(f"  Mean CV R²   : {cv_rf.mean():.3f}")

# Feature importance plot
importances = pd.DataFrame({
    "Feature":    features4,
    "Importance": rf.feature_importances_
}).sort_values("Importance", ascending=True)

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(importances["Feature"],
               importances["Importance"] * 100,
               color="#1D9E75", alpha=0.85,
               edgecolor="white", height=0.55)
for bar, val in zip(bars, importances["Importance"] * 100):
    ax.text(val + 0.3,
            bar.get_y() + bar.get_height()/2,
            f"{val:.1f}%", va="center", fontsize=9, fontweight="bold")
ax.set_xlabel("Feature Importance (%)", fontsize=11)
ax.set_title("Random Forest — Feature Importance for Revenue Growth\n"
             "(which factors drive FMCG growth the most?)",
             fontsize=12, fontweight="bold", pad=10)
ax.spines[["top","right"]].set_visible(False)
plt.tight_layout()
plt.savefig("output/ml_rf_importance.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# PCA: VISUALIZE ALL COMPANIES IN 2D

# ── Reduce 6 financial metrics → 2 principal components ──────────────
pca_features = ["EBITDA_Pct","PAT_Pct","ROE","ROA",
                "Material_Intensity","YoY_Growth"]

df_pca = master[["Company","FY"] + pca_features].dropna()
X_pca  = StandardScaler().fit_transform(df_pca[pca_features])

pca    = PCA(n_components=2)
coords = pca.fit_transform(X_pca)

df_pca["PC1"] = coords[:, 0]
df_pca["PC2"] = coords[:, 1]

explained = pca.explained_variance_ratio_ * 100

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: scatter in PCA space
ax = axes[0]
for company in ["ITC","HUL","Nestle","Dabur"]:
    sub = df_pca[df_pca["Company"] == company]
    ax.scatter(sub["PC1"], sub["PC2"],
               color=COLORS[company], s=160, zorder=5,
               label=company,
               edgecolors="white", linewidth=1)
    for _, row in sub.iterrows():
        ax.annotate(row["FY"],
                    (row["PC1"], row["PC2"]),
                    textcoords="offset points",
                    xytext=(6, 4), fontsize=8,
                    color=COLORS[company])
ax.axhline(0, color="#ccc", linewidth=0.8)
ax.axvline(0, color="#ccc", linewidth=0.8)
ax.set_xlabel(f"PC1 ({explained[0]:.1f}% variance)", fontsize=11)
ax.set_ylabel(f"PC2 ({explained[1]:.1f}% variance)", fontsize=11)
ax.set_title("PCA — Company Financial Positioning",
             fontsize=11, pad=8)
ax.legend(fontsize=10)
ax.spines[["top","right"]].set_visible(False)

# Right: loadings — how each feature drives PC1/PC2
ax2 = axes[1]
loadings = pd.DataFrame(
    pca.components_.T,
    columns=["PC1","PC2"],
    index=pca_features
)
x_pos = np.arange(len(pca_features))
width = 0.35
ax2.bar(x_pos - width/2, loadings["PC1"],
        width, color="#534AB7", alpha=0.8,
        label="PC1", edgecolor="white")
ax2.bar(x_pos + width/2, loadings["PC2"],
        width, color="#D85A30", alpha=0.8,
        label="PC2", edgecolor="white")
ax2.set_xticks(x_pos)
ax2.set_xticklabels(pca_features, rotation=30,
                    ha="right", fontsize=9)
ax2.axhline(0, color="#999", linewidth=0.8)
ax2.set_ylabel("Loading", fontsize=11)
ax2.set_title("PCA Loadings — What each PC captures",
              fontsize=11, pad=8)
ax2.legend(fontsize=10)
ax2.spines[["top","right"]].set_visible(False)

plt.suptitle("Principal Component Analysis — FMCG Financial Profiles",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("output/ml_pca.png", dpi=150, bbox_inches="tight")
plt.show()
```
